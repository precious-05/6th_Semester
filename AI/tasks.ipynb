### **BFS (Breadth 1st search)**

```python
# deque means double-ended queue.
# It comes from Python’s collections module and is specially designed for:
# Fast insertion from left
# Fast removal from left
# Fast insertion from right
# Fast removal from right
```

```python
from collections import deque   #deque means double ended queues

def bfs(graph, start_node):
    visited= set()  # Keeping the track of visited nodes
    queue = deque([start_node])     # “deque([start]) means we are creating a queue object with the start node already inside it, and it will be modified in the next steps.”
    visited.add(start_node) # adding 'A' to visited
    
    while queue:  #until the queue does not become empty
        node = queue.popleft() #current node 
        print(node, end= " ")
        
        for neighbour in graph[node]:
            if neighbour not in visited:
                visited.add(neighbour)
                queue.append(neighbour)
            
    
    

# 1st of all declare graph dictionary by adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

bfs(graph,'A')   # Sending the dictionary and start node in the bfs method
```

### **Boggle Board**

```python
class BoggleSolver:
    def __init__(self, board, dictionary):
        """
        Initialize Boggle solver with board and dictionary
        Board: 2D list of characters
        Dictionary: list of valid words
        """
        self.board = board
        self.rows = len(board)
        self.cols = len(board[0])
        self.dictionary = set(dictionary)  # O(1) lookup
        self.result = set()  # Unique words
        
        # All 8 directions: North, South, West, East, and 4 diagonals
        self.directions = [
            (-1, 0),  # North
            (1, 0),   # South
            (0, -1),  # West
            (0, 1),   # East
            (-1, -1), # North-West
            (-1, 1),  # North-East
            (1, -1),  # South-West
            (1, 1)    # South-East
        ]
    
    def is_valid_cell(self, row, col):
        """Check if cell is inside board boundaries"""
        return 0 <= row < self.rows and 0 <= col < self.cols
    
    def dfs(self, row, col, visited, current_word):
        """
        BACKEND: DFS Function - Stack ke andar kya chal raha hai
        
        visited: Set of tuples {(r1,c1), (r2,c2)} - Stack ke andar konse cells hain
        current_word: String - Stack ke andar jo cells hain unke letters ka concatenation
        
        Example:
        Stack = [(0,1), (1,1), (2,2)] 
        visited = {(0,1), (1,1), (2,2)}
        current_word = "SAN"
        """
        
        # Step 1: Current cell ko visit karo
        visited.add((row, col))
        current_word += self.board[row][col]
        
        print(f"Stack = {list(visited)} → Word = '{current_word}'")
        
        # Step 2: Check if current word is in dictionary
        if len(current_word) >= 3 and current_word in self.dictionary:
            print(f"  FOUND: '{current_word}'")
            self.result.add(current_word)
            # Dictionary se word REMOVE nahi karte!
            #  Sirf result mein add karte hain
        
        # Step 3: Prefix Check - Kya yeh word kisi dictionary word ka start hai?
        is_prefix = False
        for word in self.dictionary:
            if word.startswith(current_word):
                is_prefix = True
                break
        
        if is_prefix:
            print(f"   Prefix valid: '{current_word}' → aage badho")
            # Step 4: Saare neighbors explore karo
            for dr, dc in self.directions:
                new_row, new_col = row + dr, col + dc
                
                # Check if neighbor exists and not visited
                if self.is_valid_cell(new_row, new_col) and (new_row, new_col) not in visited:
                    print(f"    Going to neighbor ({new_row},{new_col}) = '{self.board[new_row][new_col]}'")
                    self.dfs(new_row, new_col, visited, current_word)
                    # BACKTRACK: Yahan wapas aaye
                    print(f"    Backtrack to ({row},{col}) = '{self.board[row][col]}'")
        else:
            print(f"  Prefix invalid: '{current_word}' → BACKTRACK!")
        
        # Step 5: BACKTRACK - Current cell ko visited se hatao
        visited.remove((row, col))
        # BACKEND: visited remove karne ka matlab stack se pop karna
        # current_word automatically change ho jayega because yeh string copy hai
    
    def solve(self):
        """Har cell se DFS start karo"""
        print("=" * 60)
        print("STARTING BOGGLE SOLVER...")
        print("=" * 60)
        
        for r in range(self.rows):
            for c in range(self.cols):
                print(f"\n Starting from cell ({r},{c}) = '{self.board[r][c]}'")
                print("-" * 40)
                self.dfs(r, c, set(), "")
        
        print("\n" + "=" * 60)
        print(f"FINAL RESULT: {sorted(self.result)}")
        print("=" * 60)
        return sorted(self.result)


# ============================================
# MAIN PROGRAM
# ============================================

# Board define karo
board = [
    ['M', 'S', 'E', 'F'],
    ['R', 'A', 'T', 'D'],
    ['L', 'O', 'N', 'E'],
    ['K', 'A', 'F', 'B']
]

# Dictionary define karo
dictionary = ["START", "NOTE", "SAND", "STONED"]

# Boggle solver banao
solver = BoggleSolver(board, dictionary)

# Solve karo
result = solver.solve()

print(f"\n Words found: {result}")
```

```python
def boggle_solver(board, dictionary):
    # ============================================
    # STEP 1: INITIALIZATION - Setup karo
    # ============================================
    
    # Board ki height aur width nikal lo
    total_rows = len(board)        # 4 rows
    total_cols = len(board[0])     # 4 columns
    
    # Dictionary ko set mein convert karo taake lookup fast ho
    dictionary_words = set(dictionary)  # {'START', 'NOTE', 'SAND', 'STONED'}
    
    # Jo words milenge unhe store karo (duplicate avoid karne ke liye set use karo)
    found_words = set()
    
    # ============================================
    # STEP 2: DIRECTIONS - 8 neighbors ke liye
    # ============================================
    # (row_change, column_change)
    directions = [
        (-1, 0),  # North     (up)
        (1, 0),   # South     (down)
        (0, -1),  # West      (left)
        (0, 1),   # East      (right)
        (-1, -1), # North-West (up-left)
        (-1, 1),  # North-East (up-right)
        (1, -1),  # South-West (down-left)
        (1, 1)    # South-East (down-right)
    ]
    
    # ============================================
    # STEP 3: HELPER FUNCTION - Check if cell exists
    # ============================================
    def is_cell_within_board(row, col):
        """Check karo ke di gayi cell board ke andar hai ya nahi"""
        return 0 <= row < total_rows and 0 <= col < total_cols
    
    # ============================================
    # STEP 4: MAIN DFS FUNCTION - Yahan magic hoti hai!
    # ============================================
    def dfs(current_row, current_col, visited_cells, current_word):
        """
        current_row, current_col: Abhi kaunsi cell par ho
        visited_cells: Stack - ab tak kaunsi cells visit ki hain
        current_word: Stack ke cells ke letters ka concatenation
        """
        
        # --------------------------------------------------
        # STEP 4.1: PUSH TO STACK - Current cell ko visit karo
        # --------------------------------------------------
        # visited_cells = stack = {(0,1), (1,1), (2,2)} etc
        visited_cells.add((current_row, current_col))
        
        # Current cell ka letter lo aur word mein jodo
        # board[0][1] = 'S', board[1][1] = 'A', board[2][2] = 'N'
        current_cell_letter = board[current_row][current_col]
        current_word = current_word + current_cell_letter
        
        # DEBUG: Dekho stack aur word kya hai
        # print(f"Stack = {visited_cells} → Word = '{current_word}'")
        
        # --------------------------------------------------
        # STEP 4.2: DICTIONARY CHECK - Word mila kya?
        # --------------------------------------------------
        word_length = len(current_word)
        if word_length >= 3 and current_word in dictionary_words:
            # Word mil gaya! Result mein add karo
            found_words.add(current_word)
            #  Dictionary se word REMOVE mat karo!
            #  Sirf result mein add karo
            # print(f"  FOUND: '{current_word}'")
        
        # --------------------------------------------------
        # STEP 4.3: PREFIX CHECK - Kya aage badhna chahiye?
        # --------------------------------------------------
        # Check karo ke koi dictionary word is current_word se start hota hai?
        is_prefix_valid = False
        for any_word in dictionary_words:
            if any_word.startswith(current_word):
                is_prefix_valid = True
                break
        
        # --------------------------------------------------
        # STEP 4.4: EXPLORE NEIGHBORS - Agar prefix valid hai
        # --------------------------------------------------
        if is_prefix_valid:
            # Saare 8 directions mein dekho
            for row_change, col_change in directions:
                # Neighbor ki coordinates calculate karo
                neighbor_row = current_row + row_change
                neighbor_col = current_col + col_change
                
                # Kya yeh neighbor exist karta hai?
                if is_cell_within_board(neighbor_row, neighbor_col):
                    # Kya yeh neighbor pehle visit ho chuka hai?
                    if (neighbor_row, neighbor_col) not in visited_cells:
                        # Nahi ho chuka → DFS call karo!
                        # print(f"    Going to ({neighbor_row},{neighbor_col}) = '{board[neighbor_row][neighbor_col]}'")
                        dfs(neighbor_row, neighbor_col, visited_cells, current_word)
                        # Yahan BACKTRACK ho raha hai!
                        # print(f"   Backtrack to ({current_row},{current_col}) = '{board[current_row][current_col]}'")
        # else:
            # Prefix invalid hai → yahan se BACKTRACK hoga automatically
            # print(f" Prefix invalid: '{current_word}' → BACKTRACK!")
        
        # --------------------------------------------------
        # STEP 4.5: BACKTRACK - Stack se pop karo
        # --------------------------------------------------
        # visited_cells.remove = stack se pop
        visited_cells.remove((current_row, current_col))
        # Ab stack mein pichli cell hogi, aur word bhi automatically kam ho jayega
        # kyunki current_word string copy hai, reference nahi
    
    # ============================================
    # STEP 5: START DFS FROM EVERY CELL
    # ============================================
    # Har cell ko starting point banao
    for row in range(total_rows):
        for col in range(total_cols):
            # print(f"\n📌 Starting from cell ({row},{col}) = '{board[row][col]}'")
            # Empty set = empty stack
            # Empty string = no word yet
            dfs(row, col, set(), "")
    
    # ============================================
    # STEP 6: RETURN RESULT - Sorted order mein
    # ============================================
    return sorted(found_words)


# ============================================
# MAIN PROGRAM - Code run karo
# ============================================

# 4x4 Boggle board define karo
boggle_board = [
    ['M', 'S', 'E', 'F'],
    ['R', 'A', 'T', 'D'],
    ['L', 'O', 'N', 'E'],
    ['K', 'A', 'F', 'B']
]

# Dictionary define karo - valid words
valid_dictionary = ["START", "NOTE", "SAND", "STONED"]

# Function call karo
words_found = boggle_solver(boggle_board, valid_dictionary)

# Result print karo
print("`" * 50)
print("BOGGLE BOARD:")
print("`" * 50)
for row in boggle_board:
    print("   ".join(row))

print("\n" + "_" * 50)
print()
print("DICTIONARY:", valid_dictionary)
print("_" * 50)
print("\n WORDS FOUND:", words_found)
print("_" * 50)
```
